```python
import pandas as pd
import ipywidgets as widgets

# Load the dataset
dataset = pd.read_csv('crop and fertilizer dataset.csv')

# Convert column names to string
dataset.columns = dataset.columns.astype(str)

# Define the available cities and soil types
cities = dataset['District_Name'].unique()
soil_types = dataset['Soil_color'].unique()

# Create a function to get the mean values for a given city and soil type
def get_mean_values(city, soil_type):
    mask = (dataset['District_Name'] == city) & (dataset['Soil_color'] == soil_type)
    subset = dataset[mask]
    mean_nitrogen = subset['Nitrogen'].mean()
    mean_phosphorus = subset['Phosphorus'].mean()
    mean_potassium = subset['Potassium'].mean()
    mean_rainfall = subset['Rainfall'].mean()
    mean_ph = subset['pH'].mean()
    return mean_nitrogen, mean_phosphorus, mean_potassium, mean_rainfall, mean_ph

# Create a function to predict the crop and fertilizer
def predict_crop_and_fertilizer(city, soil_type):
    mean_nitrogen, mean_phosphorus, mean_potassium, mean_rainfall, mean_ph = get_mean_values(city, soil_type)
    mask = (
        (dataset['District_Name'] == city) &
        (dataset['Soil_color'] == soil_type) &
        (dataset['Nitrogen'] >= mean_nitrogen) &
        (dataset['Phosphorus'] >= mean_phosphorus) &
        (dataset['Potassium'] >= mean_potassium) &
        (dataset['Rainfall'] >= mean_rainfall) &
        (dataset['pH'] >= mean_ph)
    )
    subset = dataset[mask]
    if len(subset) > 0:
        crop = subset['Crop'].values[0]
        fertilizer = subset['Fertilizer'].values[0]
        nitrogen = subset['Nitrogen'].values[0]
        phosphorus = subset['Phosphorus'].values[0]
        potassium = subset['Potassium'].values[0]
        rainfall = subset['Rainfall'].values[0]
        ph = subset['pH'].values[0]
    else:
        crop = 'Unknown'
        fertilizer = 'Unknown'
        nitrogen = 'Unknown'
        phosphorus = 'Unknown'
        potassium = 'Unknown'
        rainfall = 'Unknown'
        ph = 'Unknown'
    return crop, fertilizer, nitrogen, phosphorus, potassium, rainfall, ph

# Create interactive widgets for city and soil type selection
city_widget = widgets.Dropdown(options=cities, description='City:')
soil_widget = widgets.Dropdown(options=soil_types, description='Soil Type:')
output_widget = widgets.Output()

# Create a function to handle widget value changes
def on_change(change):
    if change['name'] == 'value':
        city = city_widget.value
        soil_type = soil_widget.value
        crop, fertilizer, nitrogen, phosphorus, potassium, rainfall, ph = predict_crop_and_fertilizer(city, soil_type)
        with output_widget:
            output_widget.clear_output()
            print('Predicted Crop:', crop)
            print('Predicted Fertilizer:', fertilizer)
            print('Nitrogen:', nitrogen)
            print('Phosphorus:', phosphorus)
            print('Potassium:', potassium)
            print('Rainfall:', rainfall)
            print('pH:', ph)

# Register the widget value change handlers
city_widget.observe(on_change)
soil_widget.observe(on_change)

# Display the widgets and output
display(widgets.VBox([city_widget, soil_widget]), output_widget)
```
